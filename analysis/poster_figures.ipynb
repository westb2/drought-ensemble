```python
spinup_run = pd.DataFrame(spinup_run)

def add_columns(run, baseline_run):
    run["additional_water_required_to_recover"] = baseline_run["total_storage"] - run["total_storage"]
    run["root_zone_water_required_to_recover"] = baseline_run["root_zone_storage"] - run["root_zone_storage"]
    run["shallow_water_required_to_recover"] = baseline_run["shallow_storage"] - run["shallow_storage"]
    run["deep_water_required_to_recover"] = baseline_run["deep_storage"] - run["deep_storage"]

    run["storage_delta"] = baseline_run["subsurface_storage"] - run["subsurface_storage"]
    run["storage_delta_from_baseline"] = -run["additional_water_required_to_recover"]

    run["flow_reduction"] = run["outlet_flow"] - baseline_run["outlet_flow"]
    run["percent_change_in_flow"] = (run["outlet_flow"] - baseline_run["outlet_flow"]) / baseline_run["outlet_flow"] * 100
    run["percent_change_in_flow"]  = run["percent_change_in_flow"].replace([np.inf, -np.inf, np.nan], 0)
    return run

spinup_run = pd.DataFrame(add_columns(spinup_run, spinup_run))

for run in drought_runs.values():
    run = pd.DataFrame(add_columns(run, drought_runs["baseline"]))

for run in recovery_runs.values():
    run = pd.DataFrame(add_columns(run, recovery_runs["baseline"]))
```

```python
# create an array with the timesteps for each run
SPINUP_TIMESTEP_SIZE = 24*5
DROUGHT_AND_RECOVERY_TIMESTEP_SIZE = 24*5
HOURS_IN_YEAR = 8760

spinup_timesteps = np.arange(0, spinup_run.shape[0] * SPINUP_TIMESTEP_SIZE, SPINUP_TIMESTEP_SIZE)/ HOURS_IN_YEAR 
drought_timesteps = np.arange(8760*3.0, drought_runs["baseline"].shape[0] * DROUGHT_AND_RECOVERY_TIMESTEP_SIZE + 8760.*3.0, DROUGHT_AND_RECOVERY_TIMESTEP_SIZE)/ HOURS_IN_YEAR
recovery_timesteps = np.arange(8760.*6., recovery_runs["baseline"].shape[0] * DROUGHT_AND_RECOVERY_TIMESTEP_SIZE+8760.*6., DROUGHT_AND_RECOVERY_TIMESTEP_SIZE)/ HOURS_IN_YEAR
```

```python
spinup_run["time"] = spinup_timesteps
spinup_run.set_index("time", inplace=True)
for run in drought_runs.values():
    run["time"] = drought_timesteps
    run.set_index("time", inplace=True)
for run in recovery_runs.values():
    run["time"] = recovery_timesteps
    run.set_index("time", inplace=True)

spinup_run

# baseline_run = spinup_run + drought_runs["baseline"] + recovery_runs["baseline"]
# baseline_run
```
