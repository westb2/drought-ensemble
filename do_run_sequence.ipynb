```python
def run_sequence_to_string(run_sequence):
    year_types = [run["year_type"] for run in run_sequence]
    pumping_rates = [run["pumping_rate_fraction"] for run in run_sequence]
    irrigation_flags = [run["irrigation"] for run in run_sequence]
    year_types_str = "".join(year_types)
    pumping_rates_str = "_".join([str(rate) for rate in pumping_rates])
    irrigation_str = "" + "".join(["T" if flag else "F" for flag in irrigation_flags])
    return "_".join([year_types_str, pumping_rates_str, irrigation_str])
```

```python
import hashlib

def hash_run_sequence_string(run_sequence_string):
    return hashlib.sha256(run_sequence_string.encode()).hexdigest()

def get_run_folder(domain_name, run_sequence_string):
    # TODO Change this when moving to a different computer
    ROOT = "/Users/ben/Documents/GitHub/drought-ensemble/domains/outputs"
    run_sequence_hash = hash_run_sequence_string(run_sequence_string)
    return f"{ROOT}/{domain_name}/{run_sequence_hash}"
```

```python
def do_run_sequence(domain_name, run_sequence):
    #TODO don't hardcode this
    years = {
        DRY: 2006,
        AVERAGE: 2007,
    }
    for index, run in enumerate(run_sequence):
        run_sequence_subset = run_sequence[:index + 1]
        run_sequence_str = run_sequence_to_string(run_sequence_subset)
        # check if the run folder already exists
        run_folder = get_run_folder(domain_name, run_sequence_subset)
        # If it exists, skip running this sequence
        if os.path.exists(run_folder):
            print(f"Skipping run {index + 1} for {domain_name}, folder already exists: {run_folder}")
            continue
        # Otherwise, run the model for this sequence
        else:
            print(f"Running model for {domain_name} with sequence: {run_sequence_str}")
            IC_PRESSURE_FILE = None
            run_year = years[run["year_type"]]
            pumping_rate_fraction = run["pumping_rate_fraction"]
            domain_name = f"{domain_name}"
            if index > 0:
                # If this is not the first run, use the last run's output as the initial condition
                previous_run_sequence = run_sequence[:index]
                previous_run_folder = get_run_folder(domain_name, previous_run_sequence)
                IC_PRESSURE_FILE = os.path.join(previous_run_folder, "run.out.press.00073.pfb")
                # check if the initial pressure file exists
                if not os.path.exists(IC_PRESSURE_FILE):
                    print(f"Initial pressure file does not exist: {IC_PRESSURE_FILE}. Stopping run early.")
            do_run(
                domain_name=f"{domain_name}_{run_year}",
                pumping_rate_fraction=pumping_rate_fraction,
                irrigation=run["irrigation"],
                run_dir=run_folder,
                initial_pressure_file=IC_PRESSURE_FILE
            )
    run_folder = get_run_folder(domain_name, run_sequence)
    #save run sequence to json file
    json.dump(run_sequence, open(os.path.join(run_folder, "run_sequence.json"), "w"))
    
    # Here you would typically call the function that runs the model
    # For example: run_model(run_folder)
    print(f"Running sequence for {domain_name} in folder {run_folder}")

# do_run_sequence("WOLF", run_sequence)
```
